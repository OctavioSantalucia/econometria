## üß≤ **Estructuras de control de flujo** 

### ‚ûñ Flujo

El flujo es una forma de entender la sucesi√≥n de las instrucciones de un programa, estas instrucciones se ejecutan una despu√©s de otras de forma ordenada y suelen tener el objetivo final de manipular informaci√≥n.

image.png

Sin embargo, para manipular datos no es suficiente con realizar c√°lculos o resolver expresiones, **necesitamos que de alguna forma nuestro programa pueda elegir, que sepa c√≥mo actuar en funci√≥n de determinadas situaciones, o incluso repetir una tarea si es necesario**. Para estas situaciones, existen las sentencias de control de flujo.

### ‚ûñ Sentencias de control

Se dividen en dos tipos, las de control condicional y las de control iterativo. (A las siguientes im√°genes se le denomina diagrama de flujo)
-	**Condicionales:** para elegir entre distintas posibilidades.
-	**Iterativas:** Para repetir un bloque de instrucciones.

### ‚ûñ IF Condicional

**Las sentencias de control condicionales son aquellas que nos permiten evaluar si una o m√°s condiciones se cumplen, para decir qu√© acci√≥n vamos a ejecutar.** La evaluaci√≥n de condiciones s√≥lo puede arrojar 1 de 2 resultados: True o False (verdadero o falso).

Para describir la evaluaci√≥n a realizar sobre una condici√≥n, se utilizan los operadores relacionales (==, !=, >, <, etc). Y, para evaluar m√°s de una condici√≥n simult√°neamente se utilizan los operadores l√≥gicos (not, and, or).


*Sintaxis del if condicional*

```python
if condici√≥n1:
‚ÄÇ‚ÄÇ‚ÄÇ‚ÄÇbloque c√≥digo
elif condici√≥n2:
‚ÄÇ‚ÄÇ‚ÄÇ‚ÄÇbloque c√≥digo
    ‚Ä¶
else:
‚ÄÇ‚ÄÇ‚ÄÇ‚ÄÇbloque c√≥digo
```

Eval√∫a la expresi√≥n l√≥gica **condici√≥n1** y ejecuta el primer bloque de c√≥digo si es **True**; si no, eval√∫a la siguientes condiciones **elif** hasta llegar a la primera que es **True** y ejecuta el bloque de c√≥digo asociado. Si ninguna condici√≥n es **True** ejecuta el bloque de c√≥digo despu√©s de **else**:.

````python
#Ejemplo If
persona1= "Ramona"
persona2= "Clara"
persona3= "Susana"

if persona1=="Susana": #Uso un operador l√≥gico (==) para comparar si se es igual a Susana el nombre
    print("Hola Susana te estamos llamando queremos jugar")
    print("Susana es la persona 1")
elif persona2=="Susana": 
    print("Hola Susana te estamos llamando queremos jugar")
    print("Susana es la persona 2")
elif persona3=="Susana": 
    print("Hola Susana te estamos llamando queremos jugar")
    print("Susana es la persona 3")
else:
    print("Susana no est√°")
````

üî• En Python definimos un bloque de c√≥digo despu√©s de una sentencia de control escribiendo dos puntitos : y todo el bloque que se ejecutar√° debe de estar indentado. **Luego de los dos puntitos hay que usar al menos 1 espacio pero se acostumbra a usar 4 espacios**  üî•

````python
#Bloque no indentado
if 5>4:
print("es mayor a 4")

#Nos devuelve un error
````

````python
#Bloque indentado con 4 espacios
if 5>4:
    print("es mayor a 4")
````

### ‚ûñ While Iterativo 

**Se basa en repetir un bloque de c√≥digo a partir de evaluar una condici√≥n l√≥gica, siempre que √©sta sea True** (al igual que la sentencia if). 

Nosotros como programadores debemos decidir el momento en que la condici√≥n cambie a False para hacer que el While finalice su ejecuci√≥n y as√≠ salir de la iteraci√≥n, de lo contrario estar√≠amos frente a un bucle infinito.


*Sintaxis del bucle while*

```python
while condici√≥n:
‚ÄÇ‚ÄÇ‚ÄÇ‚ÄÇbloque c√≥digo
```

Repite la ejecuci√≥n del bloque de c√≥digo mientras la expresi√≥n l√≥gica condici√≥n sea cierta. En el caso de que siempre sea cierta el bloque se ejecutar√° para siempre



````python
#Ejemplo while infinito 
while True:
    print("Es un bucle infinito")
#RECOMENDACI√ìN NO EJECUTAR POR MUCHO TIEMPO
````

````python
#Ejemplo while con condici√≥n
number=9
while number!=0: #Mientras que el n√∫mero sea distinto a cero
    print(number) #Lo imprimo
    number = number - 1 #Le resto 1
````

üî• Al igual que el IF debe estar indentado con 4 espacios  üî•

***Break en while***

La instrucci√≥n break le **proporciona la oportunidad de cerrar un bucle cuando se activa una condici√≥n externa.** Debe poner la instrucci√≥n break dentro del bloque de c√≥digo bajo la instrucci√≥n de su bucle, generalmente despu√©s de una sentencia if condicional.

````python
# Pregunta al usuario por un n√∫mero hasta que introduce 0.
while True:
    num = int(input('Introduce un n√∫mero: ')) #La funci√≥n input pide que el usuario pase un valor escrito y ponga enter
    if num == 0:
        print("Has descubierto el n√∫mero secreto")
        break
````

### ‚ûñ For Iterativo 

**Se utiliza para recorrer los elementos de un objeto iterable** (lista, tupla‚Ä¶) y ejecutar un bloque de c√≥digo, o sea, tiene un n√∫mero predeterminado de veces que itera.

En cada paso de la iteraci√≥n se tiene en cuenta a un √∫nico elemento del objeto iterable, sobre el cu√°l se pueden aplicar una serie de operaciones.


*Sintaxis del bucle for*

```python
for i in secuencia:
‚ÄÇ‚ÄÇ‚ÄÇ‚ÄÇbloque c√≥digo
```

Repite la ejecuci√≥n del bloque de c√≥digo para cada elemento de la secuencia **secuencia**, asignado dicho elemento a **i** en cada repetici√≥n.

Se puede interrumpir en cualquier momento la ejecuci√≥n del bloque de c√≥digo con la instrucci√≥n **break** o saltar la ejecuci√≥n para un determinado elemento de la secuencia con la instrucci√≥n **continue**.

````python
#Ejemplo for simple
lista=[1, 2, 3, 4]
for value in lista:
     print(value)
````

üî• Al igual que el IF y el While debe estar indentado con 4 espacios  üî•

````python
#Ejemplo for cuadrado de n√∫meros
lista=[1, 2, 3, 4]
for value in lista:
     print(value**2)
````

````python
#Ejemplo for con break 
lista=[1, 2, 3, 4]
for value in lista:
     if value==3:
       break #El break corta el ciclo for en el momento que se cumple la condici√≥n
     else:
       print(value)
````

````python
#Ejemplo for con continue
lista=[1, 2, 3, 4]
for value in lista:
     if value==3:
       continue  #El continue ignora el valor de value en el momento que se cumple la condici√≥n
     else:
       print(value)
````

***Range***

La funci√≥n range(inicio, fin-1, paso) en Python sirve para generar una secuencia de n√∫meros enteros. Se puede usar con uno, dos o tres argumentos.

- Con un solo argumento, range(n) genera una secuencia de n√∫meros del 0 al n-1.
- Con dos argumentos, range(a, b) genera una secuencia de n√∫meros desde a hasta b-1.
- Con tres argumentos, range(a, b, c) genera una secuencia desde a hasta el n√∫mero m√°s cercano a b (pero menor que b) aumentando en c en cada paso.


Es una funci√≥n muy √∫til y se utiliza com√∫nmente en ciclos for para generar una secuencia de n√∫meros sobre los que iterar.

````python
#Range con un argumento
for i in range(5):
    print(i)
````

````python
#Range con dos argumentos
for i in range(2, 8):
    print(i)
````

````python
#Range con tres argumentos
for i in range(1, 10, 2):
    print(i)
````
