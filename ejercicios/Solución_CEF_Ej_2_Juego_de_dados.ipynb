```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Primero simulemos un juego de 10 lanzamientos de dado


```python
# Simulamos una tirada de dado 10 veces
tiradas = np.random.randint(1, 7, size=10)
print(tiradas)
```

```python
# Sumamos los resultados de las tiradas y calculamos el retorno
retorno = tiradas.sum()* 1000
print(retorno)
```

```python
# metemos todo el procedimiento en una funcion de manera de simplificar el codigo
def retorno():
  tiradas = np.random.randint(1, 7, size=10)
  retorno=tiradas.sum()* 1000
  return retorno

```

```python
retorno()
```

```python
retorno()
```

`for` loop \\

El `for` en general sirve para iterar un procedimiento (sobre una lista de valores) o repetirlo. Veamos la sintaxis

```python
for i in range(3):
  print(retorno())
```

Una manera facil de meter los resultados de la repeticion en una lista es usando la siguiente sintaxis. Notar que se define la iteracion dentro de una lista (dentro de corchetes) y primero se indica el resultado

```python
[retorno() for i in range(3)]
```

Ahora simulamos este juego 1000 veces



```python
lista=[retorno() for i in range(1000)]
lista
```

Guardamos los resultados en un dataframe

```python
df=pd.DataFrame(lista,columns=["data"])
```

```python
df
```

```python
df['data'].describe()
```

```python
df['data'].hist(bins=30, edgecolor='black')
plt.title('Histograma de Retornos')
plt.xlabel('Retorno')
plt.ylabel('Frecuencia')
plt.show()
```

```python
menores_25k = sum(df['data'] < 25000)
print("Cantidad de retornos menores a 0:",menores_25k)
probabilidad_menor_25k = menores_25k / 1000 *100
print("Probabilidad de obtener un retorno menor a 25 mil pesos (en porcentaje):", probabilidad_menor_25k)

```
